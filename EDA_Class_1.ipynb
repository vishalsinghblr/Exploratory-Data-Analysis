```python
df_country=pd.read_excel('/content/drive/MyDrive/EDA Dataset/Zomatodataset/Country-Code.xlsx')
```

```python
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
#To check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3], autopct='%.2f%%')
```

Observation:
1. Zomato maximum transacions are from India, followed by USA and then UK

```python
ratings=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head()
```

Observation
1. Rating above 4.5, it is Excellent
2. Rating between 4.0 to 4.5, it is Very Good and so on
3. Many people have not given the rating

```python
plt.rcParams['figure.figsize']=(12,6)
```

```python
sns.scatterplot(data=ratings, x="Aggregate rating", y="Rating Count", hue='Rating color', palette=['black','red','orange','yellow','green','green'])
```

```python
sns.barplot(data=ratings, x="Aggregate rating", y="Rating Count", hue='Rating color', palette=['black','red','orange','yellow','green','green'])
```
